```python
movies.dropna(inplace = True)
```

```python
movies.reset_index(drop=True, inplace=True)
```

```python
movies.info()
```

```python
movies['type'] = movies['type'].apply(lambda x: x.split(','))
```

```python
movies['director'] = movies['director'].apply(lambda x: x.split(','))
```

```python
movies['cast'] = movies['cast'].apply(lambda x: x.split(','))
```

```python
movies['listed_in'] = movies['listed_in'].apply(lambda x: x.split(','))
```

```python
movies['description'] = movies['description'].apply(lambda x: x.split())
```

```python
movies
```

```python
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ','')for i in x])
```

```python
movies['director'] = movies['director'].apply(lambda x:[i.replace(' ','')for i in x])
```

```python
movies['listed_in'] = movies['listed_in'].apply(lambda x:[i.replace(' ','')for i in x])
```

```python
movies['tags'] = movies['description'] + movies['listed_in']  + movies['director'] + movies['cast'] + movies['type']
```

```python
movies_df = movies[['title','tags']]
```

```python
movies_df['tags'] = movies_df['tags'].apply(lambda x:" ".join(x))
```

```python
movies_df['tags'][0]
```

```python
movies_df['tags'] = movies_df['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=7000,stop_words='english')
```

```python
vectors = cv.fit_transform(movies_df['tags']).toarray()
```

```python
vectors[0].shape
```

```python
vectors[0]
```

```python
cv.get_feature_names_out 
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):

    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
    

```

```python
stem("a small fishing village must procure a local doctor to secure a lucrative business contract. when unlikely candidate and big city doctor paul lewis lands in their lap for a trial residence, the townsfolk rally together to charm him into staying. as the doctor's time in the village winds to a close, acting mayor murray french has no choice but to pull out all the stops. comedy drama donmckellar brendangleeson taylorkitsch gordonpinsent movie"
)
```

```python
movies_df['tags'] = movies_df['tags'].apply(stem)
```

```python
movies_df['tags'][0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
 def recommend(movie):
     movie_index = movies_df[movies_df['title'] == movie].index[0]
     distance = similarity[movie_index]
     movie_list = sorted(list(enumerate(distance)),reverse = True,key = lambda x: x[1])[1:6]
     
     for i in movie_list:
         print(movies_df.iloc[i[0]].title)

     return 
```
